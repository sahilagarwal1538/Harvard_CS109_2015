## Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

**The 3 ways of using matplotlib's pyplot:**
* (Most common way) Using global functions/'templates' (i.e. matplotlib used as a global state machine) (pick a design from available templates rather than being able to design one)
* Obtain populated Axes & its Figure container using a plot method from the pandas data analysis library, which you can 'pretty-up' 
* (More programmatic approach) Obtain an **empty** figure from global factory & then 'build' the plot using methods of Figure & Axes classes
---
*We shall focus on the 3rd approach, but just have a look at the 1st approach*

### Using matplotlib as a global state machine

##### 1. Get data 

```python
# we shall fake a monthly time series

x= pd.period_range('1980-01-01',periods=450,freq='M')
x= x.to_timestamp().to_pydatetime()

y= np.random.randn(len(x)).cumsum()
```

##### 2. Plot, Pretty-it-up, Save and display

```python
## Plot
plt.plot(x,y,label='abcd')

## Pretty-it-up
plt.gcf().set_size_inches(8,3)
plt.title('Dummy data set')
plt.xlabel('Date (in months)')
plt.ylabel('Y')
plt.grid(True)
plt.legend(loc='best',framealpha=0.5,prop={'size':'small'})
plt.tight_layout(pad=1)

## Save figure and display
plt.savefig('image1.png')
plt.show()
plt.close()
```
